```python
dummies = pd.get_dummies(df.salary, prefix = 'salary', dtype = int)
dummies
```

```python
df2 = pd.concat([df,dummies],axis = 'columns')
df2
```

```python
x = df2.drop(['last_evaluation', 'number_project','time_spend_company','Department', 'salary','left'], axis = 'columns')
x
```

```python
y = df.left
y
```

```python
from sklearn.model_selection import train_test_split
```
